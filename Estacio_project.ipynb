```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd

weather_data = pd.read_csv("weather_data.csv")

keywords = pd.read_csv("keywords.csv")

day_number = np.linspace(1, 90, 90)

# Unicode for degree symbol
a = '\u00b0'

plt.plot(day_number, weather_data["temp"])
plt.title("Metro Manila Temperature Over 90 Days")
plt.xlabel("Day")
plt.ylabel("Mean Temperature (" + str(a) + "C)")
plt.show()

x_train, x_test, y_train, y_test = train_test_split(day_number, keywords["heat"], train_size = 0.5, test_size = 0.5
                                                    #, random_state = 42
                                                    )
# Split data set into training, test, and cross-validation
# Random state was used to help with debugging
deg = np.arange(0, 10)
# Array for degrees

for n in range(10):
  best_fit_coeff = np.polyfit(x_train, y_train, deg[n])
  # Function to get coefficients of best fit function from training set
  best_fit = np.poly1d(best_fit_coeff)
  # Function to get best fit function
  plt.plot(day_number, best_fit(day_number), label = "Degree " + str(n), alpha = 0.1 + 0.1*n)
best_heat = best_fit
plt.plot(day_number, keywords["heat"])
plt.title("Best Fit Functions of Keyword: heat")
plt.xlabel("Day")
plt.ylabel("Interest over time")
plt.legend(loc = "center left")
plt.show()
print()


x_train, x_test, y_train, y_test = train_test_split(day_number, keywords["hot"], train_size = 0.5, test_size = 0.5
                                                    #, random_state = 42
                                                    )
# Split data set into training, test, and cross-validation
# Random state was used to help with debugging
deg = np.arange(0, 10)
# Array for degrees

for n in range(10):
  best_fit_coeff = np.polyfit(x_train, y_train, deg[n])
  # Function to get coefficients of best fit function from training set
  best_fit = np.poly1d(best_fit_coeff)
  # Function to get best fit function
  plt.plot(day_number, best_fit(day_number), label = "Degree " + str(n), alpha = 0.1 + 0.1*n)
best_hot = best_fit
plt.plot(day_number, keywords["hot"])
plt.title("Best Fit Functions of Keyword: hot")
plt.xlabel("Day")
plt.ylabel("Interest over time")
plt.legend(loc = "center left")
plt.show()
print()


x_train, x_test, y_train, y_test = train_test_split(day_number, keywords["weather"], train_size = 0.5, test_size = 0.5
                                                    #, random_state = 42
                                                    )
# Split data set into training, test, and cross-validation
# Random state was used to help with debugging
deg = np.arange(0, 10)
# Array for degrees

for n in range(10):
  best_fit_coeff = np.polyfit(x_train, y_train, deg[n])
  # Function to get coefficients of best fit function from training set
  best_fit = np.poly1d(best_fit_coeff)
  # Function to get best fit function
  plt.plot(day_number, best_fit(day_number), label = "Degree " + str(n), alpha = 0.1 + 0.1*n)
best_weather = best_fit
plt.plot(day_number, keywords["weather"])
plt.title("Best Fit Functions of Keyword: weather")
plt.xlabel("Day")
plt.ylabel("Interest over time")
plt.legend(loc = "center left")
plt.show()
print()


x_train, x_test, y_train, y_test = train_test_split(day_number, keywords["heat index"], train_size = 0.5, test_size = 0.5
                                                    #, random_state = 42
                                                    )
# Split data set into training, test, and cross-validation
# Random state was used to help with debugging
deg = np.arange(0, 10)
# Array for degrees

for n in range(10):
  best_fit_coeff = np.polyfit(x_train, y_train, deg[n])
  # Function to get coefficients of best fit function from training set
  best_fit = np.poly1d(best_fit_coeff)
  # Function to get best fit function
  plt.plot(day_number, best_fit(day_number), label = "Degree " + str(n), alpha = 0.1 + 0.1*n)
best_heat_index = best_fit
plt.plot(day_number, keywords["heat index"])
plt.title("Best Fit Functions of Keyword: heat index")
plt.xlabel("Day")
plt.ylabel("Interest over time")
plt.legend(loc = "center left")
plt.show()
print()


x_train, x_test, y_train, y_test = train_test_split(day_number, keywords["temperature"], train_size = 0.5, test_size = 0.5
                                                    #, random_state = 42
                                                    )
# Split data set into training, test, and cross-validation
# Random state was used to help with debugging
deg = np.arange(0, 10)
# Array for degrees

for n in range(10):
  best_fit_coeff = np.polyfit(x_train, y_train, deg[n])
  # Function to get coefficients of best fit function from training set
  best_fit = np.poly1d(best_fit_coeff)
  # Function to get best fit function
  plt.plot(day_number, best_fit(day_number), label = "Degree " + str(n), alpha = 0.1 + 0.1*n)
best_temperature = best_fit
plt.plot(day_number, keywords["temperature"])
plt.title("Best Fit Functions of Keyword: temperature")
plt.xlabel("Day")
plt.ylabel("Interest over time")
plt.legend(loc = "center left")
plt.show()
print()


x_train, x_test, y_train, y_test = train_test_split(day_number, weather_data["temp"], train_size = 0.5, test_size = 0.5
                                                    #, random_state = 42
                                                    )
# Split data set into training, test, and cross-validation
# Random state was used to help with debugging
deg = np.arange(0, 10)
# Array for degrees

for n in range(10):
  best_fit_coeff = np.polyfit(x_train, y_train, deg[n])
  # Function to get coefficients of best fit function from training set
  best_fit = np.poly1d(best_fit_coeff)
  # Function to get best fit function
  plt.plot(day_number, best_fit(day_number), label = "Degree " + str(n), alpha = 0.1 + 0.1*n)
best_weather_data = best_fit
plt.plot(day_number, weather_data["temp"])
plt.title("Best Fit Functions of Mean Temperature")
plt.xlabel("Day")
plt.ylabel("Mean Temperature (" + str(a) + "C)")
plt.legend(loc = "center left")
plt.show()
print()


plt.plot(day_number, best_weather(day_number), label = "Keyword weather")
plt.plot(day_number, best_heat(day_number), label = "Keyword heat")
plt.plot(day_number, best_hot(day_number), label = "Keyword hot")
plt.plot(day_number, best_heat_index(day_number), label = "Keyword heat index")
plt.plot(day_number, best_temperature(day_number), label = "Keyword temperature")
plt.plot(day_number, best_weather_data(day_number), label = "Mean temperature")
plt.legend(loc = "center left")
plt.title("Overlayed Degree 9 Polynomials")
plt.xlabel("Day")
plt.show()

good = day_number[best_weather_data(day_number) >= 32]
plt.scatter(good, best_weather(good), label = "Keyword weather")
plt.scatter(good, best_heat(good), label = "Keyword heat")
plt.scatter(good, best_hot(good), label = "Keyword hot")
plt.scatter(good, best_heat_index(good), label = "Keyword heat index")
plt.scatter(good, best_temperature(good), label = "Keyword temperature")
plt.scatter(good, best_weather_data(good), label = "Mean temperature")
plt.legend(loc = "center left")
plt.title("Overlayed Degree 9 Polynomials")
plt.xlabel("Day")
plt.show()
```
